```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(0.0, 5.0, 100)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)

fig, ax = plt.subplots(figsize=(5, 3))
fig.subplots_adjust(bottom=0.15, left=0.2)
ax.plot(x1, y1)
ax.set_xlabel('中文 $\mathrm{This\ is\ text}$')
ax.set_ylabel('Damped oscillation [V]')

plt.show()
```

```python
from matplotlib.font_manager import FontProperties
def get_zh_font():
    font = FontProperties()
    font.set_family('serif')
    font.set_name('Simsun')
    return font

def get_en_font():
    font = FontProperties()
    font.set_family('serif')
    font.set_name('Times New Roman')
    return font
```

```python
zh_font=get_zh_font()
en_font=get_en_font()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
fig.subplots_adjust(bottom=0.15, left=0.2)
#labels = ["小于\mathrm{1%<=10%}".replace('\\',"$\\")]
ax.plot(x1, y1, label=r"""小于$1\%<=10\%$""")
ax.set_xlabel('中文', fontproperties=zh_font)
ax.set_ylabel('Damped oscillation [V]', fontproperties=en_font)
ax.legend()
plt.show()
```

```python
import re
def get_mix_font():
    font = FontProperties()
    font.set_family('sans-serif')
    font.set_name('Simsun')
    font.set_math_fontfamily('stix')
    return font
```

```python
mix_font=get_mix_font()
```

```python
plt.rcParams['text.usetex'] = False
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
fig.subplots_adjust(bottom=0.15, left=0.2)
ax.plot(x1, y1, label=r"""小于$ \mathrm{1%<=10%}$""")
ax.set_xlabel('中文 This is a Text', fontproperties=mix_font)
ax.set_ylabel('Damped oscillation [V]', fontproperties=mix_font)
ax.legend(prop=mix_font)
plt.show()
```
